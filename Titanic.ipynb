```python
# 1. Cabin ko drop karo aur naya feature banao
data['Has_Cabin'] = data['Cabin'].notna().astype(int)
data.drop('Cabin', axis=1, inplace=True)
```

```python
data
```

```python
# 2. Age ko Pclass ke hisaab se Median se bharo (Smart way)
data['Age'] = data.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
data
```

```python
# 3. Embarked ko mode se bharo
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Kya Auratein Zyada Bachi?')
plt.show()
# Output aayega: Female ~0.74, Male ~0.19
```

```python
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Kya First Class walo ko bachaya gaya?')
plt.show()
# Output: P1=0.63, P2=0.47, P3=0.24
```

```python
# Bacche (Age < 10) ki survival alag se dekho
data['Child'] = data['Age'] < 10  # Naya column
sns.barplot(x='Child', y='Survived', data=data)
plt.title('Kya Bacche (Under 10) bache?')
plt.show()
# Output: Child = 0.58, Adult = 0.36 (Bacche zyada bache!)
```

```python
data['Title'] = data['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
sns.barplot(x='Title', y='Survived', data=data)
plt.xticks(rotation=45)
plt.show()
```

```python
data['Family_Size'] = data['SibSp'] + data['Parch'] + 1
# Dekho: Family size 1 (Akela) ki survival 35%, Family size 2-3 ki survival 50%+, Family size 5+ ki survival sirf 10% (pure family dub gayi).
sns.barplot(x='Family_Size', y='Survived', data=data)
```

```python
data['Fare_Per_Person'] = data['Fare'] / data['Family_Size']
```

```python
numeric_data = data.select_dtypes(include=[np.number])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')  # 'numeric_df' ki jagah 'numeric_data'
plt.show()
```
