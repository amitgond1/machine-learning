```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,67,37,40,55]

# line chart
plt.plot(x,y)

# display chart jp
plt.show()
```

```python
# customize chart
```

```python
# chart size
plt.figure(figsize=(6,4))

plt.plot(x, y, color='r', marker='o', linestyle='dashed',linewidth=2, markersize=12)

# heading and x-axis and y-axis
plt.title("new data ")
plt.xlabel("x-axis ")
plt.ylabel("y-axis ")


plt.show()
```

```python
# advance line chart
```

```python
x = [1,2,3,4,5]
y1 = [10,67,37,40,55]
y2 = [20,77,40,45,65]


plt.plot(x, y1 , label='sales 2024')
plt.plot(x, y2 , label='sales 2025')

plt.title("your sales")
plt.xlabel('month')
plt.ylabel('sales')

plt.legend()

plt.show()
```

```python
# bar chart
```

```python
x = ['A','B','C','D','E']
y = [10,67,37,40,55]

plt.bar(x,y)
plt.title('Bar chart')
plt.show()
```

```python
# histogram
```

```python
# data 
import random
data = [random.randint(1,10) for _ in range(100)]
# distribution ke liye use hota hai 
# histogram continues graph

plt.hist(data,bins=5)
plt.title("Histogram ")
plt.show()
```

```python
# pie chart
```

```python
# used to show part to whole relationship data

categories = ['A','B','C','D','E']
sales = [10,67,37,40,55]


plt.pie(sales,labels=categories,autopct='%1.1f%%')
plt.title("pie chart")

plt.show()
```

```python
# scatter plot
```

```python
# used to find relationship between variable

y1 = [10,67,37,40,55]
y2 = [20,77,40,45,65]

plt.scatter(y1,y2)
plt.title("scatter")

plt.show()
```

```python
# subplot
```

```python
# used to show multiple chart in on figure

# Data 1 bar chart
categories = ['Mon','Tue','Wed','Thu','Fri']
sales = [10,67,37,40,55]

# data 2 scatter plot

y1 = [10,67,37,40,55]
y2 = [20,77,40,45,65]

plt.figure(figsize=(10,4))

# first plot - bar chart
plt.subplot(1,2,1) # row, column,position
plt.bar(categories,sales)
plt.title("daily sales")
plt.xlabel("Week Days")
plt.ylabel("sales")



# second plot scatter chart
plt.subplot(1,2,2) # row, column,position
plt.scatter(y1,y2)
plt.title("user example")
plt.xlabel("user1")
plt.ylabel("user2")

plt.show()

```

```python
import pandas as pd
```

```python
data = {
    'Month':['Jan','Feb','Mar','Apr'],
    'Sales' : [12000,11000,13000,26000]
}

df = pd.DataFrame(data)
df
```

```python
# save plot
```

```python
plt.bar(df['Month'], df['Sales'])
plt.title("Monthly sales")
plt.xlabel("Month")
plt.ylabel("sales")

plt.savefig("Monthly_sales.png")

plt.show()
```
